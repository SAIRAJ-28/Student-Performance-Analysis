```python
# Importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load DATASET 

data=pd.read_csv("Student_performance_dataset.csv")
data.head(25)
```

```python
# To find DATA TYPES OF COLUMNs

print(data.dtypes)
print("\n")
print(data.shape)
```

```python
# Check Missing Values

data.isnull().sum()
```

```python
# Check Dataset Info

data.info()
```

```python
# Basic Statistics 

data.describe()
```

```python
# To find Average Scores 

data[['math_score','reading_score','writing_score','science_score']].mean()
```

```python
# Analysis by Gender Performance

Max_performance=data.groupby('gender')['total_score'].max()
Max_performance
```

```python
Min_performance=data.groupby('gender')['total_score'].min()
Min_performance
```

```python
# Test Preparation Impact 
# Students who completed test preparation scored higher.

print("0 -Not Prepared ,1-Prepared")
data.groupby('test_preparation_course')['total_score'].mean()

```

```python
# Attendence Impact
print("1 = attended,0 = absent")
data.groupby('attendence')['total_score'].count()
```

```python
# Top Performing Students
data.sort_values(by='total_score', ascending=False).head(15)
```

```python
plt.hist(data['total_score'])
plt.title("Total Score Distribution")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
#Student Performance 

subjects = ['math_score','reading_score','writing_score','science_score']

data[subjects].mean().plot(kind='bar')
plt.title("Average Score by Subject")
plt.show()
```

```python
# Correlation Anlysis

corr = data.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.show()
```
